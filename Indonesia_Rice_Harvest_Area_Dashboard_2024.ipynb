## **Data Cleaning**

**Hapus Baris Header Ganda**

```python
import pandas as pd

df = pd.read_excel("Luas Padi Indonesia 2024.xlsx")

new_columns = [
    "Provinsi",
    "Kode_Kab",
    "Kabupaten_Kota",
    "Januari",
    "Februari",
    "Maret",
    "April",
    "Mei",
    "Juni",
    "Juli",
    "Agustus",
    "September",
    "Oktober",
    "November",
    "Desember",
    "Total"
]

df.columns = new_columns

df.head()
```

```python
df = df.iloc[1:].reset_index(drop=True)
```

**Cek Missing Value**

```python
df.isnull().sum()
```

**Hapus Baris Kosong**

```python
df[df["Kabupaten_Kota"].isnull()]
```

```python
df = df.iloc[:-2]
```

**Pisahkan Data Nasional**

```python
total_indonesia = df["Total"].sum()
print(total_indonesia)
```

```python
df.info()
```

```python
bulan_cols = [
    "Januari","Februari","Maret","April","Mei","Juni",
    "Juli","Agustus","September","Oktober","November",
    "Desember","Total"
]

df[bulan_cols] = df[bulan_cols].apply(pd.to_numeric, errors='coerce')

df.info()
```

# **Exploratory Data Analysis (EDA)**

**Dataset Overview**

```python
print("Jumlah Provinsi :", df["Provinsi"].nunique())
print("Jumlah Kabupaten/Kota :", df["Kabupaten_Kota"].nunique())
print("Total Luas Panen :", df["Total"].sum())
```

**Top 10 Provinces by Harvest Area**

```python
import matplotlib.pyplot as plt

top_prov = (
    df.groupby("Provinsi")["Total"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,5))
top_prov.plot(kind="bar")
plt.title("Top 10 Provinces by Harvest Area")
plt.ylabel("Harvest Area (Ha)")
plt.show()
```

**Top 10 Regencies/Cities**

```python
top_kab = (
    df[["Kabupaten_Kota","Provinsi","Total"]]
    .sort_values("Total", ascending=False)
    .head(10)
)

top_kab
```

**Monthly Harvest Trend**

```python
bulan_cols = [
    "Januari","Februari","Maret","April","Mei","Juni",
    "Juli","Agustus","September","Oktober",
    "November","Desember"
]

monthly = df[bulan_cols].sum()

plt.figure(figsize=(12,5))
monthly.plot(marker='o')
plt.title("Monthly Harvest Area Trend")
plt.ylabel("Harvest Area (Ha)")
plt.show()
```

**Province Contribution**

```python
top5 = (
    df.groupby("Provinsi")["Total"]
    .sum()
    .sort_values(ascending=False)
    .head(5)
)

plt.figure(figsize=(7,7))
plt.pie(top5, labels=top5.index, autopct='%1.1f%%')
plt.title("Top 5 Province Contribution")
plt.show()
```

**Churn Berdasarkan Negara**

```python
df_long = pd.melt(
    df,
    id_vars=["Provinsi","Kode_Kab","Kabupaten_Kota"],
    value_vars=bulan_cols,
    var_name="Bulan",
    value_name="Luas_Panen"
)
```

```python
df_long.to_csv(
    "rice_harvest_long.csv",
    index=False
)
```
